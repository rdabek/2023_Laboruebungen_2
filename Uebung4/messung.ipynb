```python
import pandas as pd
import csv
import matplotlib.pyplot as plt

data = pd.read_csv('Ablenkung_B_Feld.csv', sep=';')

spannung = data['Spannung']
strom = data['Strom']

print(data)

# Plot
plt.figure()
plt.title("Ablenkung im $B$-Feld")
x = range(len(data['Spannung']))
plt.xlabel('Spannung $U_{b}$ gemessen [$V$]')
plt.ylabel('Strom $I_{s}^{2}$ gemessen [$A^{2}$]')
plt.errorbar(spannung, strom, fmt='.')
# plt.show()
plt.savefig('Ablenkung_im_B_Feld.png')
```
